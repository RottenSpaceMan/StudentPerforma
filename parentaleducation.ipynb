```python
from __future__ import print_function
import os
data_path = ['data']
import pandas as pd

filepath = os.sep.join(data_path + ['StudentsPerformance.csv'])
data = pd.read_csv('StudentsPerformance.csv')
```

```python
data.drop(['race/ethnicity'], axis=1, inplace=True)
```

```python
from sklearn.preprocessing import LabelBinarizer,LabelEncoder

lb = LabelBinarizer()
le = LabelEncoder()

for col in ['gender', 'lunch','test preparation course']:
    data[col] = lb.fit_transform(data[col])
for col in ['parental level of education']:
    data[col] = le.fit_transform(data[col])
```

```python
x_cols = [x for x in data.columns if x != 'parental level of education']
X_data = data[x_cols]
y_data = data['parental level of education']
```

```python
from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier(n_neighbors=3)

knn = knn.fit(X_data, y_data)

y_pred = knn.predict(X_data)
```

```python
def accuracy(real, predict):
    return sum(y_data == y_pred) / float(real.shape[0])
print(accuracy(y_data, y_pred))
```

```python
# Fit the K-nearest neighbors model with different values of k
# Store the accuracy measurement for each k

score_list = list()

for k in range(1, 21):
    
    knn = KNeighborsClassifier(n_neighbors=k)
    knn = knn.fit(X_data, y_data)
    
    y_pred = knn.predict(X_data)
    score = accuracy(y_data, y_pred)
    
    score_list.append((k, score))
    
score_df = pd.DataFrame(score_list, columns=['k', 'accuracy'])
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
sns.set_context('talk')
sns.set_style('ticks')
sns.set_palette('dark')

ax = score_df.set_index('k').plot()

ax.set(xlabel='k', ylabel='accuracy')
ax.set_xticks(range(1, 21));
```
